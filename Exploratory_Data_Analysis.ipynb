#### 5.- ANÁLISIS EXPLORATORIO DE DATOS CONTENIDOS EN INFORME

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Informe_mediciones_puntos.xlsx')
```

#### identificar NaN y posibles incidencias

```python
#Identificar posibles NaN

df.info()
```

##### Estudio de incidencia en punto de medicion TAMANGO

```python
df[df['TAMANGO_066_B1_IUN'].isna()] #Al ser pocos registros, el problema se identifica directamente con el output. Parece un error en el punto de medicion el dia 2 de julio entre las 8:30 y las 18:00
```

```python
#Comprobamos que antes y después el punto arroja mediciones
df[(df['mes'] == 7) & (df['dia'] == 2) & (df['TAMANGO_066_B1_IUN'].notnull())]
```

```python
# Filtrar datos para el día 2 de julio (mes = 7 y dia = 2)
filtered_df = df[(df['mes'] == 7) & (df['dia'] == 2)]

# Agrupar por hora y sumar la producción
hourly_production = filtered_df.groupby('hora')['TAMANGO_066_B1_IUN'].sum().reset_index()


plt.figure(figsize=(10, 6))
plt.plot(hourly_production['hora'], hourly_production['TAMANGO_066_B1_IUN'], marker='o', linestyle='-', color='b')
plt.xlabel('Hora')
plt.ylabel('MWh')
plt.title('Prod/Hora TAMANGO - 02/07 - INCIDENCIA')
plt.xticks(range(24))
plt.grid(True)
plt.show()
```

#### Estudio de variables numericas 

```python
df.describe()
```

```python

plants = ['CURICO_013_PMGD7_LCH', 'MALLOCO_012_PMGD5_PNF', 'RENGO_015_PMGD5_SLM', 'TAMANGO_066_B1_IUN', 'URIBE_023_PMGD2_UAS']

# Crear un DataFrame vacío para almacenar la produccion media por hora y planta
hourly_activity = pd.DataFrame()

# Calcular la media de produccion por hora para cada planta
for plant in plants:
    plant_data = df.groupby('hora')[plant].mean().reset_index()
    plant_data.rename(columns={plant: 'Produccion'}, inplace=True)
    plant_data['planta'] = plant
    hourly_activity = pd.concat([hourly_activity, plant_data])

# Crear gráficos por planta
plants = hourly_activity['planta'].unique()
for plant in plants:
    plt.figure(figsize=(12, 6))
    plant_data = hourly_activity[hourly_activity['planta'] == plant]
    sns.lineplot(data=plant_data, x='hora', y='Produccion', marker='o')
    plt.xlabel('Hora')
    plt.ylabel('Produccion Media (MWh)')
    plt.title(f'Produccion Media por Hora - {plant}')
    plt.grid(True)
    plt.xticks(range(24))
    plt.show()
```

```python
# junto
custom_labels = ['CURICO', 'MALLOCO', 'RENGO', 'TAMANGO', 'URIBE']
plt.figure(figsize=(12, 6))
hourly_avg = df.groupby('hora')[plants].mean()
hourly_avg.plot()
plt.title('Produccion Media por Hora')
plt.xlabel('Hora')
plt.ylabel('Produccion Media (MWh)')
plt.legend(custom_labels, loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Media de producción mensual para cada planta
monthly_avg = df.groupby(['mes'])[['CURICO_013_PMGD7_LCH', 'MALLOCO_012_PMGD5_PNF', 'RENGO_015_PMGD5_SLM', 'TAMANGO_066_B1_IUN', 'URIBE_023_PMGD2_UAS']].mean()

plt.figure(figsize=(12, 6))
monthly_avg.plot()
plt.title('Producción Media por Mes')
plt.xlabel('Mes')
plt.ylabel('Producción Media')
plt.legend(['CURICO', 'MALLOCO', 'RENGO', 'TAMANGO', 'URIBE'], loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

```python
# Visualización de outliers y desviación
plt.figure(figsize=(14, 8))
for i, plant in enumerate(['CURICO_013_PMGD7_LCH', 'MALLOCO_012_PMGD5_PNF', 'RENGO_015_PMGD5_SLM', 'TAMANGO_066_B1_IUN', 'URIBE_023_PMGD2_UAS'], 1):
    plt.subplot(2, 3, i)
    sns.boxplot(df[plant])
    plt.title(plant)
    plt.xlabel('Producción')
plt.tight_layout()
plt.show() 
```
